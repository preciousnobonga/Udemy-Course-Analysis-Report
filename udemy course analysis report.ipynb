```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# set plot style
sns.set(style='whitegrid')
```

```python
# load dataset (from https://www.kaggle.com/datasets/andrewmvd/udemy-courses)
df = pd.read_csv(r"C:\Users\Precious Nobonga\Downloads\udemy courses\udemy_courses.csv")
df.head(20)
```

```python
# check dataset structure and types
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# check for duplicates
df.duplicated()
```

```python
# drop unnecessary columns
df.drop(columns=['course_id', 'url'], inplace=True, errors='ignore')
df
```

```python
# Total Subscribers by Subject
subject_subscribers = df.groupby('subject')['num_subscribers'].sum().reset_index()
subject_subscribers = subject_subscribers.sort_values(by='num_subscribers', ascending=False)
subject_subscribers
```

```python
# Plot Total Subscribers by Subject
plt.figure(figsize=(10, 6))
sns.barplot(x='num_subscribers', y='subject', data=subject_subscribers, palette='viridis', hue='subject', legend=False)
plt.title('Total Subscribers by Subject', fontsize=16)
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Subject', fontsize=12)
plt.show()
```

```python
#  Total Subscribers by Course
top_courses = df[['course_title', 'num_subscribers']].sort_values(by='num_subscribers', ascending=False).head(10)
top_courses
```

```python
# Plot top 10 courses by number of subscribers

plt.figure(figsize=(10,6))
sns.barplot(x='num_subscribers', y='course_title', data=top_courses, palette='coolwarm', hue='course_title', legend=False)
plt.title('Top 10 Courses by Number of Subscribers', fontsize=16)
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Course Title', fontsize=12)
plt.show()
```

```python
# Total subscribers by course level

level_subscribers = df.groupby('level')['num_subscribers'].sum().reset_index()
level_subscribers['percentage'] = (level_subscribers['num_subscribers'] / level_subscribers['num_subscribers'].sum()) * 100
level_subscribers
```

```python
# plot total subscribers by course level

plt.figure(figsize=(8, 6))
sns.barplot(x='num_subscribers', y='level', data=level_subscribers, palette='magma', hue='level', legend=False)
plt.title('Total Subscribers by Course Level', fontsize=16)
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Level', fontsize=12)
plt.show()
```

```python
# Engagement levels by subject

subject_reviews = df.groupby('subject')['num_reviews'].sum().reset_index()
subject_reviews = subject_reviews.sort_values(by='num_reviews', ascending=False)
subject_reviews
```

```python
# Plot Total Reviews by Subject

plt.figure(figsize=(10, 6))
sns.barplot(x='num_reviews', y='subject', data=subject_reviews, palette='Blues_d', hue='subject', legend=False)
plt.title('Total Reviews by Subject', fontsize=16)
plt.xlabel('Number of Reviews', fontsize=12)
plt.ylabel('Subject', fontsize=12)
plt.show()
```
